```python
investment_data_2 = investment_data.drop(['2010', '2009', '2008',
       '2007', '2006', '2005', '2004', '2003', '2002', '2001', '2000', '1999',
       '1998'], axis=1)
```

```python
## filter only retail investment data
retail_investment_data = investment_data_2[(investment_data_2['DataSeries'] == '    Wholesale & Retail Trade') |
      (investment_data_2['DataSeries'] == '        Wholesale Trade') |
      (investment_data_2['DataSeries'] == '        Retail Trade')]
```

```python
retail_investment_data.head(10)
```

```python
## reshape the data for better analysis

df_retail_final = (
    retail_investment_data
    .set_index("DataSeries")   # lock series as column headers
    .T                         # transpose years into rows
    .reset_index()             # bring years back as a column
    .rename(columns={"index": "Year"})
)

df_retail_final.head(10)
```

```python
## Plotting the retail investment data of line chart over the years
plt.figure(figsize=(12,6))
sns.lineplot(data=df_retail_final, x='Year', y='    Wholesale & Retail Trade', marker='o', label='Wholesale & Retail Trade')
sns.lineplot(data=df_retail_final, x='Year', y='        Wholesale Trade', marker='o', label='Wholesale Trade')
sns.lineplot(data=df_retail_final, x='Year', y='        Retail Trade', marker='o', label='Retail Trade')
plt.title('Foreign Direct Investment in Retail Sector (2011-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Investment Amount (in million SGD)', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
## average income of individuals by year
average_income_by_year = individual_income_data_2.groupby('year_of_assessment')['amount'].mean().reset_index()
average_income_by_year.rename(columns={'amount': 'average_income'}, inplace=True)
average_income_by_year.head(10)
```

```python
## filter only for 2024 

individual_income_2024 = individual_income_data_2[individual_income_data_2['year_of_assessment'] == 2024]
individual_income_2024.head(10)
```

```python
## What is average income of individuals in 2024 by income type
average_income_2024_by_type = individual_income_2024.groupby('income_type')['amount'].mean().reset_index()
average_income_2024_by_type.rename(columns={'amount': 'average_income_2024'}, inplace=True)
average_income_2024_by_type.head(10)
```

```python
plt.figure(figsize=(15, 10))

sns.barplot(
    data=average_income_2024_by_type,
    x='average_income_2024',
    y='income_type',
    color='green'
)

plt.xscale('log')

plt.title('Average Income by Income Type (2024, Log Scale)', fontsize=16)
plt.xlabel('Average Income (log scale)', fontsize=14)
plt.ylabel('Income Type', fontsize=14)
plt.grid(axis='x', which='both')
plt.show()

```
